```python
import pandas as pd
from matplotlib import colormaps
import seaborn as sns
```

```python
HeartRate = pd.read_csv('heart_disease_uci.csv', index_col = 0)
HeartRate['Target'] = 0
for i in range(1, 921):
    if (HeartRate.num[i] == 0):
        continue
    else:
        HeartRate.Target[i] = 1
HeartRate
```

```python
HeartRate.plot.scatter(x = 'age', y = 'chol', c = 'Target', cmap = 'bwr')
```

```python
HeartRate.isna().sum()
```

```python
pd.plotting.scatter_matrix(HeartRate)
```

```python
compare_columns = ['chol', 'age', 'trestbps', 'thalch', 'oldpeak']
sns.pairplot(HeartRate, y_vars = compare_columns, x_vars = compare_columns, hue = HeartRate['Target'])
```
